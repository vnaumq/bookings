```python
top5 = bookings\
        .query("is_canceled == 0")\
        .groupby(['country'],as_index=False)\
        .aggregate({'is_canceled' : 'count'})\
        .sort_values('is_canceled', ascending= False)
top5.head(5)
#or
top5_1 = bookings.query("is_canceled == 0").value_counts('country')
top5_1.head(5)



          
```

Users from which countries have made the most successful bookings? A reservation is considered successful if it has not been canceled (is_canceled variable).

```python
count_nights = bookings\
               .groupby(['hotel'],as_index=False)\
               .agg({'stays_total_nights' : 'mean'}).round(2)
count_nights
```

```python
popular_month = bookings\
                    .query('arrival_date_year == 2016 or arrival_date_year == 2017')\
                    .groupby(['arrival_date_year','arrival_date_month'])\
                    .agg({'arrival_date_month' : 'count'})
                    
popular_month
```

Now let's analyze the dates of planned arrival (arrival_date_year).

Which month was the most frequently booked for in 2016? Did the most popular month change in 2017?

```python
year_count_canceled = bookings\
                    .query('is_canceled == 1 and hotel == "City Hotel"')\
                    .groupby(['arrival_date_year','arrival_date_month'])\
                    .agg({'arrival_date_month' : 'count'})
year_count_canceled
#or
year_count_canceled = bookings\
                    .query('is_canceled == 1 and hotel == "City Hotel"')\
                    .groupby(['arrival_date_year'])\
                    .arrival_date_month\
                    .value_counts()
year_count_canceled
```

```python
def has_kids(x):
    if x>0:
        return True
    else:
        return False    
bookings['has_kids'] = bookings.total_kids.apply(has_kids)

#or
bookings['has_kids'] = bookings.total_kids >=1
bookings[['has_kids','is_canceled']].value_counts() / bookings[['has_kids']].value_counts()*100

```

Churn rate is the percentage of subscribers (e.g., to push notifications from a website) who unsubscribed from a communication channel, abandoned the service within a certain period of time. In other words, it is the ratio of the number of users who left to the total number of users, expressed as a percentage.
